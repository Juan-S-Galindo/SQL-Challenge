```python
#To connect with SQL server 
from sqlalchemy import create_engine

#Normal import 
import pandas as pd 
import matplotlib.pyplot as plt
import numpy as np

%matplotlib inline
```

```python
import databaseconfig as cfg
# Conect to postgres engine.

engine = create_engine(f'postgresql://{cfg.mysql["user"]}:{cfg.mysql["password"]}@{cfg.mysql["host"]}:5432/{cfg.mysql["db"]}')
connection = engine.connect()
```

```python
#Test
Salary_df=pd.read_sql('Select * From salaries', connection)
Salary_df.head()
```

```python
plt.figure(figsize = (20,10))
plt.hist(Salary_df['salary'])
plt.ylabel('Frequency')
plt.xlabel('Salary Ranges')
plt.show()
```

```python
SalaryByTitle_df = pd.read_sql('SELECT titles.title, salaries.salary FROM salaries JOIN titles ON titles.emp_no = salaries.emp_no', connection)
```

```python
plt.figure(figsize = (20,10))
plt.bar(SalaryByTitle_df.groupby('title').mean().index.to_list(),SalaryByTitle_df.groupby('title').mean()['salary'].to_list())
plt.ylabel('Average Salary')
plt.xlabel('Titles')
plt.show()
```
